```python
#import dependencies
import numpy as np
import pandas as pd
```

```python
#import datasets
#encoding = 'latin-1' was used for df2 because of UTF-8 error
df1 = pd.read_csv(r"C:\Users\lenovo\Desktop\DSN\December data science project\Regression (Life Expectancy prediction)\led Average.csv")
df2 = pd.read_csv(r"C:\Users\lenovo\Desktop\DSN\December data science project\Regression (Life Expectancy prediction)\Regions.csv", encoding ='latin-1')
```

```python
#print shape of the two datasets loaded
print("df1 shape:", df1.shape)
print("df2 shape:", df2.shape)
```

```python
#prints columns names
print("df1 columns:", df1.columns)
print("df2 columns:", df2.columns)
```

```python
#print first 5 of the dataset
df1.head()
```

```python
#print the the number years and how many times they occur
df1["Year"].value_counts()
```

```python
#fill the year null value with 2016, so as to easily bring out 2016 data
df1["Year"] = df1["Year"].fillna(2016)
```

```python
#check if 2016 has been added
df1["Year"].value_counts()
```

```python
#print out only 2016 data from the dataset
df1 = df1[df1.Year ==2016]
```

```python
#view the first five(5) data
df1.head()
```

```python
#print shape of data
df1.shape
```

```python
#reset the index of the data 
df1 = df1.reset_index(drop = True)
```

```python
#print the first 5 of the second dataset
df2.head()
```

```python
#merging of the two datasets together
df = pd.merge(df1, df2, left_on = "Country", right_on = "TableName")
```

```python
#print out the final data
print("shape:", df.shape)
```

```python
#Export my new dataframe to csv for easy analysis
df.to_csv(r"C:\Users\lenovo\Desktop\DSN\December data science project\Regression (Life Expectancy prediction)\df.csv", index = False)
```
